```python
"""
2. 读入企业的基本信息df0，选取年度区间为2018-12-31的数据
并拼接基本数据与财务数据，主键为“股票代码”
"""
df0 = pd.read_excel('/Users/liuchang/PycharmProjects/New_EKG_CreditRating/EKG_Data/STK_ListedCoInfoAnl.xlsx')
```

```python
df0.info() # 一共有3607家上市公司
```

```python
# 拼接规则暂定如下，基本数据拼接偿债能力, 存储于df0_1中, 再拼接比率结构，存储于df0_2中,...以此类推，最后拼接完发展能力df6后，存储于df0_6中
df0_1 = pd.merge(df0, df1, left_on='Symbol', right_on='Stkcd', how='left')
```

```python
df0_1.info()
```

```python
df0_1.shape
```

```python
df0_3 = pd.merge(df0_2, df3, left_on='Symbol', right_on='Stkcd', how='left')
```
